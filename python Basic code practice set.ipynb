```python
# create list for 5 days snacks
print(""" 
Monday    : pizza
Teuseday  : pasta
Wednesday : vadapav
Thursday  : biryani
Friday    : parata
""")
```

```python
day=["Monday","Teuseday", "Wednesday","Thursday","Friday"]
snacks=["pizza","pasta","vadapav","biryani","parata"]
for i in range(5):
    print(day[i],  ":",snacks[i])
```

```python
for day, snacks in zip(day,snacks):
    print(day,":",snacks)
```

```python
def display_customer_info(names, mobile_numbers):
    if len(names)!=len(mobile_numbers):
        print("Error: Mobile number not matches")
        return
    for name, mobile in zip(names,mobile_numbers):
        print(f"cutomer:{name},mobile:{mobile}")
names=["Alisce","Bob","Charlie"]
mobile_numbers=["984860221","9601324562","9948456123"]
display_customer_info(names,mobile_numbers)
```
